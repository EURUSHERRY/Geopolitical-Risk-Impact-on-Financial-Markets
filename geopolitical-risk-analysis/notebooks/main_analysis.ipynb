```python
import yfinance as yf
import pandas as pd
import numpy as np  # Import numpy for logarithmic function

industry_stocks = {
    "World_Communication_Services": ["META", "GOOGL", "GOOG", "NFLX", "DIS", "VZ", "CMCSA", "T", "TMUS", "DTEGY"],
    "World_Consumer_Discretionary": ["AMZN", "TSLA", "HD", "TM", "LVMUY", "MCD", "LOW", "BKNG", "TJX", "NKE"],
    "World_Consumer_Staples": ["PG", "COST", "NSRGY", "WMT", "KO", "PEP", "PM", "UL", "LRLCY", "MDLZ"],
    "World_Energy": ["XOM", "CVX", "SHEL", "COP", "TTE", "BP", "CNQ", "SLB", "ENB", "MPC"],
    "World_Financials": ["JPM", "BRK-B", "V", "MA", "BAC", "WFC", "HSBC", "RY", "SPGI", "GS"],
    "World_Health_Care": ["LLY", "UNH", "NVO", "JNJ", "MRK", "ABBV", "TMO", "AZN", "NVS", "ABT"],
    "World_Industrials": ["GE", "CAT", "UNP", "SIEGY", "UBER", "RTX", "HON", "ETN", "SBGSY", "BA"],
    "World_Information_Technology": ["MSFT", "AAPL", "NVDA", "AVGO", "ASML", "AMD", "CRM", "ADBE", "ACN", "ORCL"],
    "World_Materials": ["LIN", "BHP", "AIQUY", "SHW", "SHECY", "RIO", "FCX", "CRH", "ECL", "GLNCY"],
    "World_Real_Estate": ["PLD", "AMT", "EQIX", "WELL", "CCI", "PSA", "DLR", "SPG", "O", "CSGP"],
    "World_Utilities": ["NEE", "SO", "IBDRY", "DUK", "CEG", "ENLAY", "NGG", "AEP", "SRE", "D"]
}

# Set uniform start and end dates
start_date = '1986-01-01'
end_date = '2023-12-31'

industry_returns = pd.DataFrame(index=pd.date_range(start=start_date, end=end_date, freq='D'))

# Calculate daily log returns for each stock and industry
for industry, stocks in industry_stocks.items():
    print(f"Calculating returns for {industry}")
    industry_data = pd.DataFrame(index=pd.date_range(start=start_date, end=end_date, freq='D'))
    for stock in stocks:
        try:
            data = yf.download(stock, start=start_date, end=end_date)
            if not data.empty:
                data['Log Return'] = np.log(data['Adj Close'] / data['Adj Close'].shift(1))  # Compute log returns
                industry_data[stock] = data['Log Return']
            else:
                print(f"No available data for {stock}")
        except Exception as e:
            print(f"Failed to calculate returns for {stock}: {e}")

    # Calculate industry returns as the mean return of all stocks in the industry
    industry_returns[industry] = industry_data.mean(axis=1)

# Calculate monthly volatility for each industry
monthly_volatility = industry_returns.resample('M').std()

industry_returns.to_excel("Industry_Daily_Log_Returns.xlsx")
monthly_volatility.to_excel("Industry_Monthly_Volatility.xlsx")

print("Daily log returns and monthly volatility calculated and saved.")
```

```python
import pandas as pd

wti_data_path = 'data_wti.xlsx'
wti_data = pd.read_excel(wti_data_path)
wti_data.set_index('Date', inplace=True)
wti_data['WTI'] = wti_data['WTI'].interpolate(method='linear')

industry_returns_path = "Industry_Daily_Log_Returns.xlsx"
industry_returns = pd.read_excel(industry_returns_path)
industry_returns.set_index('Unnamed: 0', inplace=True)

# Ensure datetime-like index
industry_returns.index = pd.to_datetime(industry_returns.index)

monthly_covariance_results = pd.DataFrame(index=industry_returns.columns)

# Calculate monthly covariance between each industry and WTI
for month, group in industry_returns.groupby(pd.Grouper(freq='MS')):
    combined_data = group.join(wti_data['WTI'])
    monthly_correlation = combined_data.corr().iloc[0, 1]  # Calculate correlation
    wti_std = combined_data['WTI'].std()  # WTI standard deviation
    for industry_column in industry_returns.columns:
        industry_std = combined_data[industry_column].std()  # Industry standard deviation
        industry_wti_covariance = monthly_correlation * industry_std * wti_std
        industry_name = 'WTI_' + industry_column  # Add prefix to industry name
        monthly_covariance_results.loc[industry_name, month] = industry_wti_covariance

# Transpose the DataFrame
monthly_covariance_results = monthly_covariance_results.T.reset_index()
monthly_covariance_results.columns.name = None

monthly_covariance_results.to_excel('Industry_WTI_Monthly_Covariance_Transposed.xlsx', index=False)

print("Monthly covariance between each industry and WTI calculated, transposed, and saved.")
```

```python
# ADF, Jarque-Bera Test
results = []

for column_name, column_data in data_ln.items():
    result = {
        'Column': column_name,
        'Mean': column_data.mean(),
        'Standard Deviation': column_data.std(),
        'Skewness': column_data.skew(),
        'Kurtosis': column_data.kurtosis(),
        'Jarque-Bera Value': jarque_bera(column_data)[0],
    }

    adf_result = adfuller(column_data.values.ravel())
    result['ADF Statistic'] = adf_result[0]

    results.append(result)

results_df = pd.DataFrame(results)

results_df = results_df.reindex(columns=['Column', 'Mean', 'Standard Deviation', 'Skewness', 'Kurtosis', 'Jarque-Bera Value', 'ADF Statistic'])

results_df
```

```python
import pandas as pd
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

base_features = ['WT', 'PT', 'MB', 'NT', 'TT', 'BW', 'EW', 'TA']
y_columns = data.columns[8:]

average_results = []

TEST_LENGTH = 91
# (456-2)*0.2 = 90.8

def smape(y_true, y_pred):
    denominator = (np.abs(y_true) + np.abs(y_pred)) / 2.0
    diff = np.abs(y_true - y_pred) / denominator
    diff[denominator == 0] = 0.0
    return 100 * np.mean(diff)

for industry_column in y_columns:
    data[industry_column + '_lag1'] = data[industry_column].shift(1)
    data[industry_column + '_lag2'] = data[industry_column].shift(2)

    X = data[base_features + [industry_column + '_lag1', industry_column + '_lag2']].dropna()
    y = data.loc[X.index, industry_column]

    X_train = X.iloc[:-TEST_LENGTH]
    X_test = X.iloc[-TEST_LENGTH:]
    y_train = y.iloc[:-TEST_LENGTH]
    y_test = y.iloc[-TEST_LENGTH:]

    svr = SVR()
    svr.fit(X_train, y_train)
    svr_pred = svr.predict(X_test)
    mse_svr = mean_squared_error(y_test, svr_pred)
    rmse_svr = np.sqrt(mse_svr)
    smape_svr = smape(y_test, svr_pred)
    r2_svr = r2_score(y_test, svr_pred)

    gbm = GradientBoostingRegressor()
    gbm.fit(X_train, y_train)
    gbm_pred = gbm.predict(X_test)
    mse_gbm = mean_squared_error(y_test, gbm_pred)
    rmse_gbm = np.sqrt(mse_gbm)
    smape_gbm = smape(y_test, gbm_pred)
    r2_gbm = r2_score(y_test, gbm_pred)

    average_results.append({
        'Industry': industry_column,
        'MSE SVR': mse_svr,
        'RMSE SVR': rmse_svr,
        'SMAPE SVR': smape_svr,
        'R-squared SVR': r2_svr,
        'MSE GBM': mse_gbm,
        'RMSE GBM': rmse_gbm,
        'SMAPE GBM': smape_gbm,
        'R-squared GBM': r2_gbm
    })

for result in average_results:
    print(result)
```
